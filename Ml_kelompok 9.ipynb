## Membuat model Word2Vec

```python
# model word2vec yang digunakan membutuhkan data untuk melakukan pelatihan
# data yang digunakan untuk pelatihan diambil dari idwiki-latest-pages-articles.xml.bz2 yang merupakan data text bahasa indonesia
import io
import time
from datetime import timedelta
import gensim
start_time = time.time()
id_wiki = gensim.corpora.WikiCorpus('idwiki-latest-pages-articles.xml.bz2', lemmatize=False, dictionary={})
article_count = 0
with io.open('idwiki.txt', 'w', encoding='utf-8') as wiki_txt:
    for text in id_wiki.get_texts():
        wiki_txt.write(" ".join(text) + '\n')
        article_count += 1
        if article_count % 10000 == 0:
            print('{} articles processed'.format(article_count))
    print('total: {} articles'.format(article_count))
finish_time = time.time()
print('Elapsed time: {}'.format(timedelta(seconds=finish_time-start_time)))
```

```python
# Membangun model word2vec dimana model akan menggunakan data latih yang telah dikumpulkan sebelumnya
import time
import multiprocessing
from datetime import timedelta
from gensim.models import word2vec
start_time = time.time()
sentences = word2vec.LineSentence('idwiki.txt')
w2v = word2vec.Word2Vec(sentences, size=200, workers=multiprocessing.cpu_count()-1)
w2v.save('ml_word2vec.model')
finish_time = time.time()
print('Finished. Elapsed time: {}'.format(timedelta(seconds=finish_time-start_time)))
```

## Membuat fungsi untuk mengganti kalimat

```python
print('fungsi dibawah akan memeriksa kata yang sudah baru dari dataset kata baku yang telah dikumpulkan sebelumnya, jika kata tidak ada pada dataset kata baku maka kata baku akan ditemukan menggunakan algoritma word2vec yang telah dibuat sebelumnya')
```

```python
# fungsi utama
def katabaru(sentence):
    tokens = sentence.split()
    lemmatized_tokens=[]
    for token in tokens:
        try:
            if(token not in k):
                x = w2v.wv.most_similar(token, topn=1)[0][0]
                lemmatized_tokens.append(x)
            else:
                x = token
                lemmatized_tokens.append(x)
        except:
            x = token
            lemmatized_tokens.append(x)
    hasil = " ".join(lemmatized_tokens)        
    return hasil
```

```python
data_uji = df['comment'][3:13]
```

```python
r = data_uji.apply(katabaru)
r
```

```python
rev = pd.DataFrame(r,columns=["comment"])
```

```python
dataakhir = pd.concat((data_uji,rev), axis=1)
dataakhir.columns = ['data lama','data baru']
```
